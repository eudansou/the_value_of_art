```python
# Convert price strings into integers
data['price'] = data['price'].str.replace('USD', '')
data['price'] = data['price'].str.replace('.', '')
data['price'] = data['price'].astype(int)
```

```python
# Convert strings of yearCreation into integers

# Remove 'circa'
data['yearCreation'] = data['yearCreation'].str.replace('circa', '')
# Remove 'Circa'
data['yearCreation'] = data['yearCreation'].str.replace('Circa', '')
# Strip empty spaces
data['yearCreation'] = data['yearCreation'].str.strip()
```
