```python
import os

image_dir = "raw_data/artDataset"

# Fazer lista com nome das imagens
images = [f for f in os.listdir(image_dir) if f.endswith(".png")]
print(len(images))
images
```

```python
data["images"] = [f'{i}' for i in range(1, len(data) + 1)]
```

```python
# numerar as imagens e colocar em ordem crescente
image_number = [img.replace('image_', '').replace('.png', '') for img in images]
image_number = sorted(image_number, key=lambda x: int(x))
print(len(image_number))
image_number
```

```python
# voltar com o nome das imagens para o nome completo "image_{i}.png", agora na lista ordenada
images = [f"image_{i}.png" for i in image_number]
print(len(images))
images
```

```python
# fazer uma lista com os preÃ§os para fazer o y
prices = data["price"]
prices = prices.values.tolist()
prices
```

```python
# Tive que fazer o pip install do open cv. Colocar no requirements?
# pip install opencv-python
import cv2

# FunÃ§Ã£o para transformar a imagem em matriz prÃ©-processando
def load_and_preprocess(image_dir, image_name, target_size=(224, 224)):
    img = cv2.imread(f"{image_dir}/{image_name}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, target_size) / 255.0   # Resize + normalize
    return img


# from tensorflow.keras.preprocessing.image import load_img, img_to_array

# def load_and_preprocess_keras(image_dir, image_name, target_size=(224, 224)):
#     img_path = f"{image_dir}/{image_name}"
#     img = load_img(img_path, target_size=target_size)  # Carrega e redimensiona
#     img = img_to_array(img) / 255.0  # Converte para array e normaliza
#     return img
```

```python
X = []
y = []

# Faz o X usando a funÃ§Ã£o anterior e o y
for image_name, price in zip(images, prices):
    img = load_and_preprocess(image_dir, image_name)
    X.append(img)
    y.append(price)

# Convertendo para arrays NumPy
X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)
```
